```python
# Task 1 RESTAURANT RATINGS
# Analyze the distribution of aggregate ratings and determine the most common rating range

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#load the dataset
df = pd.read_csv(r"C:\Users\Acer\Downloads\Dataset.csv")
df.shape
```

```python
# drop missing or zero ratings
ratings = df['Aggregate rating']
ratings = ratings[ratings > 0]
```

```python
# plot the distribution
plt.figure(figsize = (8,5))
sns.histplot(ratings, bins = 10, kde = True , color = 'red')
plt.title('Distribution of aggregate ratings')
plt.xlabel('Rating range')
plt.ylabel('Number of restaurants')
plt.grid(True)
plt.show()
```

```python
# identify the most common rating range
most_common_rating = ratings.value_counts().idxmax()
print(f"Most common rating range: {most_common_rating}")
```

```python
#calculate average number of votes received by restaurants

average_votes = df['Votes'].mean()
print(f"Average number of votes received by restaurants: {average_votes:2f}")
```
